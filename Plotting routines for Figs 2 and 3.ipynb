```python
%matplotlib inline
clf()
    
xoff=0.16
yoff=0.2
width=0.88-xoff
height=0.91-yoff
        
dashes=[3,2]
labelx=-0.13

############################################

fig_width_pt = 246.0
inches_per_pt = 1.0/72.27
golden_mean = (sqrt(5)-1.0)/2.0
fig_width = fig_width_pt*inches_per_pt
fig_height = fig_width_pt*inches_per_pt*golden_mean
fig_size =  [1.3*fig_width,1.3*fig_height]
params = {'backend': 'pdf',
    'axes.labelsize': 10,
        'font.size': 10,
            'xtick.labelsize': 8,
                'ytick.labelsize': 8,
                    'text.usetex': False, 
                      'mathtext.fontset': 'stixsans',
                        'figure.figsize': fig_size,
                            'figure.subplot.left': xoff,
                                'figure.subplot.bottom': yoff,
                                    'figure.subplot.right': xoff+width,
                                        'figure.subplot.top': yoff+height,
'mathtext.fontset': 'stixsans'}

rcParams.update(params)
rc('font',**{'family':'sans-serif','sans-serif':['Tex Gyre Heros']})

#rc('mathtext',fontset='custom')
#rc('mathtext',sf='Helvetica Neue LT Std')
#rc('mathtext',it='Helvetica Neue LT Std:italic')

figure(1)

subplots_adjust(hspace=0.2,wspace=0.0)

labx=0.03
laby=0.88

#########################################
###########################################

subplot(111)

Z=array(np.loadtxt('fig2/mirPD1FzC2t1.txt',delimiter='\t'))
X=array(np.loadtxt('fig2/mirPD1FxC2t1.txt',delimiter='\t'))
Y=array(np.loadtxt('fig2/mirPD1FyC2t1.txt',delimiter='\t'))/3e11

minline=np.loadtxt('fig2/mirPD1minC2t1.txt',delimiter='\t')
minoptline=np.loadtxt('fig2/mirPD1minC2WKt1.txt',delimiter='\t')


#plot([pow(10,e[0]) for e in minoptline],[pow(10,e[1]) for e in minoptline],marker='*',color='g',ms=5,mec='g',mfc='None',mew=1,linestyle='None')
linWK,=plot([pow(10,e[0]) for e in minoptline],[pow(10,e[1])/3e11 for e in minoptline],color='g',lw=2)
linWK.set_dashes([3,1])
lin,=plot([pow(10,e[0]) for e in minline],[pow(10,e[1])/3e11 for e in minline],color='b',lw=2)        
matplotlib.rcParams['contour.negative_linestyle'] = 'solid'

contourf(X,Y,np.log10(Z),cmap='hot',levels=np.linspace(-3,2,num=26))
CS=contour(X,Y,np.log10(Z),levels=[e for e in arange(-3.0,0.0,0.2)]+[e for e in arange(0.2,2.0,0.2)],colors='k')
CS2=contour(X,Y,np.log10(Z),levels=[0.0],colors='r',linewidths=2)

print(np.linspace(-3,2,num=26))

gca().set_xscale('log')
gca().set_yscale('log')
setp(gca(),xlim=[pow(10,-2.5),pow(10,6)],ylim=[pow(10,3.5)/3e11,pow(10,8.5)/3e11])

#setp(gca(),xlim=[1,pow(10,6)],ylim=[pow(10,4),pow(10,7)],xticklabels=[])
#matplotlib.rcParams['contour.negative_linestyle'] = 'solid'

#contourf(X,Y,Z,cmap='hot',levels=[-9,-8,-7,-6,-5,-4,-3,-2,-1])
#CS=contour(X,Y,Z,levels=[0.2,0.4,1],colors='k')

#plot([e for e in arange(0,50,0.1)],[(1+e)**0.5 for e in arange(0,50,0.1)],color='w',lw=2)
#setp(gca(),xlim=[1,50.0],ylim=[0,10],xticks=[1,10,20,30,40,50])

#clabel(CS, fmt='%2.1f',fontsize=8, inline=1,inline_spacing=3,manual=[(pow(10,-1.8),2e-6),(1.0,1e-5),(1.0,2e-4)])
#clabel(CS2, fmt='%2.1f',fontsize=8, inline=1,inline_spacing=3,manual=[(pow(10,-1),pow(10,4.5)/3e11)])
clabel(CS, fmt='%2.1f',fontsize=8, inline=1,inline_spacing=3,manual=[(pow(10,-2),pow(10,4)/3e11),(pow(10,1),pow(10,4.5)/3e11),(pow(10,1),pow(10,6.0)/3e11),(pow(10,1),pow(10,7.1)/3e11)])
clabel(CS2, fmt='%2.1f',fontsize=8, inline=1,inline_spacing=3,manual=[(pow(10,-1),pow(10,4.5)/3e11)])

#legend([lin,linWK],[r'optimum',r'optimum (WK)'],frameon=False,loc=4,handlelength=1.4)

ylabel(r'$|\sigma| = \Delta M/M_{\rm tot}$')
xlabel(r'$K_M$ [nM]')

###########################################

# savefig('FigureC2ct100.svg')
#savefig('FigureC2t1.eps')
#fixbbox('FigureC2t100.eps')
#os.system('epstopdf FigureC2t1.eps')
show()
```
